```python
from sklearn.datasets import load_iris, load_breast_cancer, load_digits, load_wine
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Ask the user to input the dataset
dataset_name = input("Enter the name of the dataset: ")

# Load the dataset
if dataset_name == "iris":
    dataset = load_iris()
elif dataset_name == "breast_cancer":
    dataset = load_breast_cancer()
elif dataset_name == "digits":
    dataset = load_digits()
elif dataset_name == "wine":
    dataset = load_wine()
else:
    print("Dataset not found")
    exit()

# Divide the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(dataset.data, dataset.target, test_size=0.2)

# Preprocess the data
# e.g., normalize the data, handle missing values, feature selection, etc.

# Train the classifiers
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)

mlp = MLPClassifier()
mlp.fit(X_train, y_train)

rf = RandomForestClassifier()
rf.fit(X_train, y_train)

svm = SVC()
svm.fit(X_train, y_train)

# Evaluate the classifiers
knn_pred = knn.predict(X_test)
mlp_pred = mlp.predict(X_test)
rf_pred = rf.predict(X_test)
svm_pred = svm.predict(X_test)

# Calculate the evaluation metrics
knn_accuracy = accuracy_score(y_test, knn_pred)
knn_precision = precision_score(y_test, knn_pred, average="macro")
knn_recall = recall_score(y_test, knn_pred, average="macro")
knn_fscore = f1_score(y_test, knn_pred, average="macro")

mlp_accuracy = accuracy_score(y_test, mlp_pred)
mlp_precision = precision_score(y_test, mlp_pred, average="macro")
mlp_recall = recall_score(y_test, mlp_pred, average="macro")
mlp_fscore = f1_score(y_test, mlp_pred, average="macro")

rf_accuracy = accuracy_score(y_test, rf_pred)
rf_precision = precision_score(y_test, rf_pred, average="macro")
rf_recall = recall_score(y_test, rf_pred, average="macro")
rf_fscore = f1_score(y_test, rf_pred, average="macro")

svm_accuracy = accuracy_score(y_test, svm_pred)
svm_precision = precision_score(y_test, svm_pred, average="macro")
svm_recall = recall_score(y_test, svm_pred, average="macro")
svm_fscore = f1_score(y_test, svm_pred, average="macro")

# Output the evaluation metrics of each classifier

#KNN

print("-------------------------------------")
print("Evaluation Metrics for KNN:")
print("Accuracy:", knn_accuracy)
print("Precision:", knn_precision)
print("Recall:", knn_recall)
print("F1 Score:", knn_fscore)
print()

#MLP

print("-------------------------------------")
print("Evaluation Metrics for MLP:")
print("Accuracy:", mlp_accuracy)
print("Precision:", mlp_precision)
print("Recall:", mlp_recall)
print("F1 Score:", mlp_fscore)
print()

# Random Forest

print("-------------------------------------")
print("Evaluation Metrics for Random Forest:")
print("Accuracy:", rf_accuracy)
print("Precision:", rf_precision)
print("Recall:", rf_recall)
print("F1 Score:", rf_fscore)
print()

#SVM

print("-------------------------------------")
print("Evaluation Metrics for SVM:")
print("Accuracy:", svm_accuracy)
print("Precision:", svm_precision)
print("Recall:", svm_recall)
print("F1 Score:", svm_fscore)
```

#### Bar Graph

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a list of evaluation metric values for each classifier
accuracy = [knn_accuracy, mlp_accuracy, rf_accuracy, svm_accuracy]
precision = [knn_precision, mlp_precision, rf_precision, svm_precision]
recall = [knn_recall, mlp_recall, rf_recall, svm_recall]
f1_score = [knn_fscore, mlp_fscore, rf_fscore, svm_fscore]

# Set the positions of the bars on the x-axis
x_pos = np.arange(len(accuracy))

# Set the width of each bar
bar_width = 0.2

# Create a figure and axis object
fig, ax = plt.subplots()

# Create the bars for each evaluation metric for each classifier
rects1 = ax.bar(x_pos, accuracy, bar_width, label='Accuracy', color='tab:blue')
rects2 = ax.bar(x_pos + bar_width, precision, bar_width, label='Precision', color='tab:green')
rects3 = ax.bar(x_pos + 2*bar_width, recall, bar_width, label='Recall', color='tab:red')
rects4 = ax.bar(x_pos + 3*bar_width, f1_score, bar_width, label='F1 Score', color='tab:orange')

# Add labels and title
ax.set_xlabel('Classifier')
ax.set_ylabel('Evaluation Metric Value')
ax.set_title('Comparison of Evaluation Metrics for Different Classifiers on '+dataset_name.capitalize()+' Dataset')
ax.set_xticks(x_pos + 1.5*bar_width)
ax.set_xticklabels(['KNN', 'MLP', 'Random Forest', 'SVM'])

# Add legend outside the graph
handles, labels = ax.get_legend_handles_labels()
fig.legend(handles, labels, loc='upper right')

# Show the plot
plt.show()
```
